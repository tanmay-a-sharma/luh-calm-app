```python
import pandas as pd
import numpy as np

smoker = pd.read_csv("smokerResponseSP24.csv")
smoker.head()
```

```python
smoker = smoker.drop(columns=['Anything else I should know?'])
smoker.head()
```

```python
smoker = smoker.drop(columns=['Timestamp'])
smoker.head()
```

```python
# df = pd.read_csv('data.csv')

# Step 2: Filter and save based on 'Role'
for role in smoker['Role'].unique():
    # Filter the DataFrame
    filtered_df = smoker[smoker['Role'] == role]

    # Step 3: Save each filtered DataFrame to a CSV file
    # The file name will be the role name, e.g., 'NI.csv', 'HOC.csv', 'Active.csv'
    filtered_df.to_csv(f'{role}.csv', index=False)
```
